```python
def get_winner_loser(df, year):
    # filter selected year
    data = df[df['year'] == year]
    
    # average weekly sales per store
    avg_sales = (
        data.groupby('store', as_index=False)['weekly_sales']
            .mean().round(2)
    )
    
    # top 5 stores (winners)
    winners = avg_sales.sort_values('weekly_sales', ascending=False).head(5)
    winners['rank'] = 'top 5'
    
    # bottom 5 stores (losers)
    losers = avg_sales.sort_values('weekly_sales', ascending=True).head(5)
    losers['rank'] = 'bottom 5'
    
    # combine
    result = pd.concat([winners, losers])
    
    return result.sort_values('weekly_sales', ascending=False)


# 🔹 usage

result_2010 = get_winner_loser(df, 2010)
result_2011 = get_winner_loser(df, 2011)
result_2012 = get_winner_loser(df, 2012)

print ("\n=====2010======")
print(result_2010)
print ("\n=====2011======")
print(result_2011)
print ("\n=====2012======")
print(result_2012)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# define temperature bins
bins = [-float('inf'), 32, 50, 68, 77, 90, float('inf')]

x_labels = [
    'Very Cold',
    'Cold',
    'Comfortable',
    'Mild/Ideal',
    'Warm',
    'Hot'
]

legend_labels = {
    'Very Cold': '<32 (very cold)',
    'Cold': '32-50 (cold)',
    'Comfortable': '50-68 (comfortable)',
    'Mild/Ideal': '68-77 (mild/ideal)',
    'Warm': '77-90 (warm)',
    'Hot': '90+ (hot)'
}

# create temperature category
df['temp_category'] = pd.cut(
    df['temperature'],
    bins=bins,
    labels=x_labels
)

# calculate avg weekly sales by temp + holiday
result = (
    df.groupby(['temp_category', 'holiday_flag'])['weekly_sales']
    .mean()
    .round(2)
    .reset_index()
)

# rename holiday labels
result['holiday_flag'] = result['holiday_flag'].map({
    True: 'Holiday',
    False: 'Non-Holiday'
})

# plot
plt.figure(figsize=(12,6))

ax = sns.barplot(
    data=result,
    x='temp_category',
    y='weekly_sales',
    hue='holiday_flag',
    palette='Set2'
)

# titles and labels
plt.title('Weekly Sales by Temperature Category and Holiday Status', fontsize=14)
plt.xlabel('Temperature Category')
plt.ylabel('Average Weekly Sales')

# add values on bars
for container in ax.containers:
    ax.bar_label(container, fmt='%.1f', padding=3)

# temperature range legend
temp_legend = "\n".join([
    f"{k}: {v}" for k, v in legend_labels.items()
])

plt.figtext(
    1.02, 0.5,
    temp_legend,
    fontsize=10,
    va='center',
    linespacing=2.8
)

plt.tight_layout()
plt.savefig("C:/Users/abc/Desktop/data/Walmart_Sales/graph/Weekly Sales by Temperature Category and Holiday Status.png",bbox_inches='tight')
plt.show()
```
